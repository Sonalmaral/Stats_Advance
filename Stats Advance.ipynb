1. What is a random variable in probability theory?
>> A random variable is a numerical outcome of a random phenomenon. It maps outcomes of a probability experiment to numbers.

2. What are the types of random variables?
>> Discrete Random Variable: Takes countable values (e.g., number of heads in 3 coin tosses).
Continuous Random Variable: Takes any value within a given range (e.g., height, weight, time).

3. What is the difference between discrete and continuous distributions?
>> Discrete distributions describe random variables with specific, countable outcomes.
Continuous distributions describe variables with infinite possible values within a range.

4. What are probability distribution functions (PDF)?
>> For discrete variables, PDF (also called PMF - Probability Mass Function) gives the probability of each possible value.

For continuous variables, PDF is a function whose area under the curve gives the probability over an interval.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
>> CDF gives the probability that a random variable is less than or equal to a certain value.

PDF/PMF gives the probability of a specific value (discrete) or density at a point (continuous).

6. What is a discrete uniform distribution?
>> A discrete uniform distribution assigns equal probability to all outcomes in a finite set.
E.g., Rolling a fair 6-sided die: P(1) = P(2) = ... = P(6) = 1/6.

7. What are the key properties of a Bernoulli distribution?
>> Only two outcomes: Success (1) or Failure (0).

Parameter: p (probability of success).

Mean = p, Variance = p(1 - p).

8. What is the binomial distribution, and how is it used in probability?
>> Describes the number of successes in n independent Bernoulli trials.

Parameters: n (trials), p (success probability).

Used in scenarios like flipping coins, defect rates, etc.

9. What is the Poisson distribution and where is it applied?
>> Models the number of events occurring in a fixed interval of time or space.

Used when events happen independently and at a constant average rate (λ).

Examples: Number of calls per hour, accidents per day.

11. What are the characteristics of a normal distribution?
>> Bell-shaped, symmetric about the mean.

Defined by mean (μ) and standard deviation (σ).

Mean = Median = Mode.

12. What is the standard normal distribution, and why is it important?
>> A normal distribution with μ = 0 and σ = 1.

Used to standardize data and compute Z-scores for comparisons and hypothesis testing.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
>> CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution.
Importance: Allows use of normal-based methods even when the original data isn’t normal.

14. How does the Central Limit Theorem relate to the normal distribution?
>> It justifies using the normal distribution to approximate the distribution of the sample mean when the sample size is large (usually n ≥ 30).

15. What is the application of Z statistics in hypothesis testing?
>> Z-statistics measure how far a sample statistic is from the population mean in standard deviations. It helps test hypotheses about means when population variance is known.

16. How do you calculate a Z-score, and what does it represent?

>> It represents how many standard deviations a value (X) is from the mean (μ).

Z= X−μ /σ

17. What are point estimates and interval estimates in statistics?
>> Point Estimate: Single value estimate of a parameter (e.g., sample mean).

Interval Estimate: A range of values (confidence interval) within which the parameter is expected to lie.

18. What is the significance of confidence intervals in statistical analysis?
>> They provide a range of plausible values for a population parameter and convey the uncertainty in the estimate.

19. What is the relationship between a Z-score and a confidence interval?
>> Z-scores determine the margin of error for confidence intervals. For example, for 95% confidence, Z ≈ 1.96.

20. How are Z-scores used to compare different distributions?
>> By converting raw scores to Z-scores, you can compare values from different normal distributions on a common scale.

21. What are the assumptions for applying the Central Limit Theorem?
>> Sample size is sufficiently large (n ≥ 30 recommended).

Observations are independent.

Finite variance in the population.

22. What is the concept of expected value in a probability distribution?
>> The expected value is the average or mean outcome of a random variable over many trials.

23. How does a probability distribution relate to the expected outcome of a random variable?
>> A probability distribution defines the likelihood of each outcome, and the expected value is the weighted average of these outcomes.
